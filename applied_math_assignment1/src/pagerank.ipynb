```python
import networkx as nx
import csv
import numpy as np

edges = []
with open("/data2/home/zhaoyi/labs/USTC-labs/applied_math_assignment1/datasets/PageRank_Dataset.csv") as csv_reader:
    csv_reader = csv.reader(csv_reader)
    for line in csv_reader:
        if line[0] != 'node_1':
            edges.append(line)

G = nx.DiGraph()
for edge in edges:
    G.add_edge(edge[0], edge[1])

pagerank_results = nx.pagerank(G, alpha=0.85)

k = 20
top_k = sorted(pagerank_results.items(), key=lambda x:x[1], reverse=True)[:k]
print("top20-id:\n" + str([x[0] for x in top_k]))
print("top20-pagerank:\n" + str([x[1] for x in top_k]))


output_data = [[0 for i in range(2)] for j in range(len(pagerank_results))]
counter = 0
for key in pagerank_results:
    output_data[counter] = [key, pagerank_results[key]]
    counter += 1
with open("/data2/home/zhaoyi/labs/USTC-labs/applied_math_assignment1/results/pagerank/PageRank_result.csv", 'w', newline='') as csv_writer:
    csv_writer = csv.writer(csv_writer)
    csv_writer.writerows(output_data)


print("--------------------------------------------")
for beta in list(np.arange(0,1.1,0.1)):
    if beta > 1:
        beta = 1.0
    pagerank_results = nx.pagerank(G, alpha=beta)
    k = 20
    top_k = sorted(pagerank_results.items(), key=lambda x: x[1], reverse=True)[:k]
    print("β="+str(beta))
    print("top20-id:\n" + str([x[0] for x in top_k]))
    print("top20-pagerank:\n" + str([x[1] for x in top_k]))
    print("--------------------------------------------\n")
```
